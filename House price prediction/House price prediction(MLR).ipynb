```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

code column to be converted to seperate column called A and B column
there are two types
oneHotEncoder
LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelencoder_x=LabelEncoder()

print(labelencoder_x.fit_transform(x[:,3]))
print(x[:,3])
```

```python
x[:,3]=labelencoder_x.fit_transform(x[:,3])

x[:10,]
```

```python
onehotencoder=OneHotEncoder(categorical_features=[3])
onehotencoder.fit_transform(x)
x = onehotencoder.fit_transform(x).toarray()
x
```

```python
x=x[:,1:]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print("Train of x: ",x_train.shape) # x is always 2d so that (40,1)
print("Train of y:",y_train.shape) # y is always 1d so that (40,)
print("Test of x:",x_test.shape)
print('test of y:',y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression() 
regressor.fit(x_train,y_train)
y_pred_test = regressor.predict(x_test)
print(y_pred_test)
y_pred_train = regressor.predict(x_train)
print(y_pred_train)
```

```python
#Metrics is used to evaluate the model , like r square,sse,mse,rmse
from sklearn.metrics import r2_score
model_score_train= r2_score(y_train, y_pred_train)
model_score_test= r2_score(y_test, y_pred_test)
print(model_score_train)
print(model_score_train.round(2))
print(model_score_test.round(2))
```
